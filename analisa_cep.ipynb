#Analisa CEP

Modelo de resposta: pd.DataFrame


Regiões: Norte, Nordeste, Sudeste, Centro-Oeste e Sul

Sub-regiões: Capital, Metropolitana, Interior

```python
import pandas as pd
def analisa_cep(lista_cep: list | tuple) -> pd.DataFrame:
  lista_cep = pd.DataFrame(lista_cep, columns=['cep'])
  return pd.DataFrame(list(lista_cep.cep.apply(dados_cep)), columns=['regiao', 'estado', 'subregiao'])

def dados_cep(cep: str) -> tuple:

  if len(cep) > 5 and cep[:5].isnumeric():
    cep = int(cep[:5])
  else:
    cep = -1

  if cep < 1000:
    return ('erro', 'erro', 'erro')
  elif (1000 <= cep <= 5999) or (8000 <= cep <= 8499):
    return ('Sudeste', 'Sao Paulo', 'Capital')
  elif (6000 <= cep <= 7999) or (8500 <= cep <= 9999) or (11000 <= cep <= 11999):
    return ('Sudeste', 'Sao Paulo', 'Metropolitana')
  elif 12000 <= cep <= 19999:
    return ('Sudeste', 'Sao Paulo', 'Interior')
  elif 20000 <= cep <= 23799:
    return ('Sudeste', 'Rio de Janeiro', 'Capital')
  elif 23799 < cep <= 26600:
    return ('Sudeste', 'Rio de Janeiro', 'Metropolitana')
  elif 26601 <= cep <= 28999:
    return ('Sudeste', 'Rio de Janeiro', 'Interior')
  elif 29000 <= cep <= 29099:
    return ('Sudeste', 'Vitoria', 'Capital')
  elif 29100 <= cep <= 29999:
    return ('Sudeste', 'Vitoria', 'Interior')
  elif 30000 <= cep <= 31999:
    return ('Sudeste', 'Minas Gerais', 'Capital')
  elif 31999 < cep <= 34999:
    return ('Sudeste', 'Minas Gerais', 'Metropolitana')
  elif 35000 <= cep <= 39999:
    return ('Sudeste', 'Minas Gerais', 'Interior')
  elif 40000 <= cep <= 41999:
    return ('Nordeste', 'Bahia', 'Capital')
  elif 41999 < cep <= 44470:
    return ('Nordeste', 'Bahia', 'Metropolitana')
  elif 44471 <= cep <= 48999:
    return ('Nordeste', 'Bahia', 'Interior')
  elif 49000 <= cep <= 49099:
    return ('Nordeste', 'Sergipe', 'Capital')
  elif 49100 <= cep <= 49999:
    return ('Nordeste', 'Sergipe', 'Interior')
  elif 50000 <= cep <= 52999:
    return ('Nordeste', 'Pernambuco', 'Capital')
  elif 52999 < cep <= 54999:
    return ('Nordeste', 'Pernambuco', 'Metropolitana')
  elif 55000 <= cep <= 56999:
    return ('Nordeste', 'Pernambuco', 'Interior')
  elif 57000 <= cep <= 57099:
    return ('Nordeste', 'Alagoas', 'Capital')
  elif 57100 <= cep <= 57999:
    return ('Nordeste', 'Alagoas', 'Interior')
  elif 58000 <= cep <= 58099:
    return ('Nordeste', 'Paraiba', 'Capital')
  elif 58100 <= cep <= 58999:
    return ('Nordeste', 'Paraiba', 'Interior')
  elif 59000 <= cep <= 59099:
    return ('Nordeste', 'Rio Grande do Norte', 'Capital')
  elif 59100 <= cep <= 59999:
    return ('Nordeste', 'Rio Grande do Norte', 'Interior')
  elif 60000 <= cep <= 60999:
    return ('Nordeste', 'Ceara', 'Capital')
  elif 60999 < cep <= 61900:
    return ('Nordeste', 'Ceara', 'Metropolitana')
  elif 61901 <= cep <= 63999:
    return ('Nordeste', 'Ceara', 'Interior')
  elif 64000 <= cep <= 64099:
    return ('Nordeste', 'Piaui', 'Capital')
  elif 64100 <= cep <= 64999:
    return ('Nordeste', 'Piaui', 'Interior')
  elif 65000 <= cep <= 65099:
    return ('Nordeste', 'Maranhao', 'Capital')
  elif 65100 <= cep <= 65999:
    return ('Nordeste', 'Maranhao', 'Interior')
  elif 66000 <= cep <= 66999:
    return ('Norte', 'Para', 'Capital')
  elif 66999 < cep <= 67999:
    return ('Norte', 'Para', 'Metropolitana')
  elif 68000 <= cep <= 68899:
    return ('Norte', 'Para', 'Interior')
  elif 68900 <= cep <= 68914:
    return ('Norte', 'Amapa', 'Capital')
  elif 68915 <= cep <= 68999:
    return ('Norte', 'Amapa', 'Interior')
  elif 69000 <= cep <= 69099:
    return ('Norte', 'Amazonas', 'Capital')
  elif 69100 <= cep <= 69299:
    return ('Norte', 'Amazonas', 'Interior')
  elif 69300 <= cep <= 69339:
    return ('Norte', 'Roraima', 'Capital')
  elif 69340 <= cep <= 69389:
    return ('Norte', 'Roraima', 'Interior')
  elif 69900 <= cep <= 69920:
    return ('Norte', 'Acre', 'Capital')
  elif 69921 <= cep <= 69999:
    return ('Norte', 'Acre', 'Interior')
  elif 70000 <= cep <= 70999:
    return ('Centro-Oeste', 'Distrito Federal', 'Capital')
  elif 71000 <= cep <= 73699:
    return ('Centro-Oeste', 'Distrito Federal', 'Interior')
  elif 73699 < cep <= 73999:
    return ('Centro-Oeste', 'Goias', 'Capital')
  elif 74895 <= cep <= 76799:
    return ('Centro-Oeste', 'Goias', 'Interior')
  elif 77000 <= cep <= 77270:
    return ('Norte', 'Tocantins', 'Capital')
  elif 77300 <= cep <= 77995:
    return ('Norte', 'Tocantins', 'Interior')
  elif 78000 <= cep <= 78109:
    return ('Centro-Oeste', 'Mato Grosso', 'Capital')
  elif 78110 <= cep <= 78899:
    return ('Centro-Oeste', 'Mato Grosso', 'Interior')
  elif 78900 <= cep <= 78930:
    return ('Centro-Oeste', 'Rondonia', 'Capital')
  elif 78931 <= cep <= 78999:
    return ('Centro-Oeste', 'Rondonia', 'Interior')
  elif 79000 <= cep <= 79129:
    return ('Centro-Oeste', 'Mato Grosso do Sul', 'Capital')
  elif 79130 <= cep <= 79999:
    return ('Centro-Oeste', 'Mato Grosso do Sul', 'Interior')
  elif 79999 <= cep <= 82999:
    return ('Sul', 'Parana', 'Capital')
  elif 82999 < cep <= 83800:
    return ('Sul', 'Parana', 'Metropolitana')
  elif 83801 <= cep <= 87999:
    return ('Sul', 'Parana', 'Interior')
  elif 88000 <= cep <= 82999:
    return ('Sul', 'Santa Catarina', 'Capital')
  elif 82999 < cep <= 88469:
    return ('Sul', 'Santa Catarina', 'Metropolitana')
  elif 88470 <= cep <= 89999:
    return ('Sul', 'Snta Catarina', 'Interior')
  elif 90000 <= cep <= 91999:
    return ('Sul', 'Rio Grande do Sul', 'Capital')
  elif 91999 < cep <= 94900:
    return ('Sul', 'Rio Grande do Sul', 'Metropolitana')
  elif 94901 <= cep <= 99999:
    return ('Sul', 'Rio Grande do Sul', 'Interior')
  else:
    return ('erro', 'erro', 'erro')
```
